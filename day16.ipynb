```python
from collections import deque
import copy
OPEN = "|O|"

DATA = puzzle.test
DATA = puzzle.data

dbg = False

class Path():
    def __init__(self, initial_valve, initial_minute):
        self.steps = initial_valve.name
        self.current = initial_valve
        self.flow = 0 #how much per tick
        self.total_flow = 0 #total so far
        self.minute = initial_minute
        self.progress = []
        self.opened = set()
        
    def add_step(self, what):
        
        if what == OPEN:
            self.steps += "<=>" + self.current.name + OPEN
            self.current.open = True
            self.flow += self.current.flow
            self.opened.add(self.current.name)
        else:
            self.steps += "<=>" + what.name
            self.current = what
        self.total_flow += self.flow
        self.minute +=1
        self.progress.append(self.flow)

    
    def __repr__(self):
        return f"P: {self.steps}  \t\tflow: {self.flow}\t  total: {self.total_flow}\t  progress: {sum(self.progress)}  "        
    
    
class Valve():
    def __init__(self, line):
        line = line.replace(" tunnels lead to valves ","").replace(" tunnel leads to valve ","").replace(", ",";").replace("Valve ","").replace(" has flow rate=",";").split(";")
        self.name= line[0]
        self.flow = int(line[1])
        self.conns= set(line[2:])
        if not self.flow:# it does not make sense to open a valve with 0 flow
            self.open = True
        else:
            self.open = False

    def connections (self, valves):
        self.conns= [ valves[name] for name in self.conns]
        return self
        
    def __repr__(self):
        return f"V: {self.name}\tflow: {self.flow}  \tconnects with: {self.conns}"
    
    def next_steps (self):
        next_valves = list(self.conns)
        if not self.open:
            next_valves.append(OPEN)
        return next_valves


valves = [Valve(data) for data in DATA]
valves = { v.name: v for v in valves}
for v in valves.values():
    v.connections(valves)
paths = deque()
minute = 0
ini = Path(valves["AA"], 1)
paths.append( ini)


import time



def solution_1(paths, minute):
    start = time.time()

    best_total = 0
    best_progress = 0
    best_path = None
    while minute <30:
        if dbg:
            print(f"\n\n===========\minute:{minute} paths:")
            for path in paths:
                print (path)        
            print("=============\n\n")

        curr_path = paths.popleft()
        minute = curr_path.minute

        #adding new steps available
        for new_step in curr_path.current.next_steps():
            new_path = copy.deepcopy(curr_path)
            new_path.add_step(new_step)
            paths.append(new_path)
            
            if new_path.total_flow > best_total:
                print (f"Minute: {minute}\t paths: {len(paths)}\t best_total: {best_total}\t best_progress: {best_progress}")
                best_total = new_path.total_flow 
                best_path = new_path
            if sum(new_path.progress) > best_progress:
                print (f"Minute: {minute}\t paths: {len(paths)}\t best_total: {best_total}\t best_progress: {best_progress}")
                best_progress = sum(new_path.progress)
            
        if not curr_path.current.name not in curr_path.opened :
            curr_path.add_step(OPEN)

        """
        if minute >8:
            fraction = 1.5
            if minute >10:
                fraction = 1.2
            if minute >18:
                fraction = 1.1
            new = deque()
            for path in paths:
                if path.total_flow < best_total/fraction:
                    continue
                new.append(path)
            paths = new"""
        
        if minute > 7:
            fraction = 1.7
            if minute >9:
                fraction = 1.6
            if minute >12:
                fraction = 1.5
            if minute >16:
                fraction = 1.4
            if minute >18:
                fraction = 1.3
            if minute >21:
                fraction = 1.2
            if minute >23:
                fraction = 1.1
            if minute >25:
                fraction = 1.05
            new = deque()
            for path in paths:
                if path.total_flow < best_total/fraction:
                    continue
                new.append(path)
            paths = new
        
            
    
    print("\n\n")
    print (f"Minute: {minute}\t paths: {len(paths)}\t best_total: {best_total}\nbest path: {best_path} ")
    """for path in paths:
        if path.flow > 78:
            print (path)"""
    
    end = time.time()
    print(f"\n\nelapsed {end - start} seconds")


solution_1(paths, minute)
```
