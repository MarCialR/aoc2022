```python
def parse_blueprints(data):
    """Blueprint 1: Each ore robot costs 4 ore. Each clay robot costs 2 ore. Each obsidian robot costs 3 ore and 14 clay. Each geode robot costs 2 ore and 7 obsidian."""
    for line in data:
        print(line)
        line = line.replace("Blueprint ","").replace(": Each ore robot costs","").replace(" ore. Each clay robot costs","").replace(" ore. Each obsidian robot costs","").replace(" ore and","").replace(" clay. Each geode robot costs","").replace("ore and","").replace("obsidian.","")
        print (line)
```

```python
def solution_1(data):
    
    cubes = set([tuple(map(int,x.split(",") ))for x in data])
    sides = [ (1,0,0),(0,1,0),(0,0,1),(-1,0,0),(0,-1,0),(0,0,-1) ]

    counter = 6*len(cubes)

    for cube in cubes:
        for side in sides:
            if  tuple(sum(t) for t in zip(cube, side)) in cubes:
                counter -=1
    return counter

assert solution_1 (puzzle.test) == 64
      
print( f"Solution 1:  {solution_1 (puzzle.data)} is the surface area of my scanned lava droplet")
```
