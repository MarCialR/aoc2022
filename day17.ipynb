```python
from collections import deque
from itertools import cycle

dbg = False

DOWN = 'âŒ„'
LEFT = '<'
RIGHT = '>'
VOID = '.'
SOLID = '#'

LINE = ([(0,3),(0,4),(0,5),(0,6)], 'LINE', 1)
PLUS = ([(0,5),(1,4),(1,5),(1,6),(2,5)], 'PLUS', 3)
VERTEX = ([(0,6),(0,5),(0,4),(1,4),(2,4)], 'VERTEX', 3)
STICK = ([(0,6),(1,6),(2,6),(3,6)], 'STICK', 4)
SQUARE = ([(0,5),(0,6),(1,5),(1,6)], 'SQUARE', 2)


class Rock():
    def __init__ (self, data, height):
        self.data = [(e[0] ,e[1]) for e in data[0]]
        self.name = data[1]
        self.size = data[2]
        self.height = 0
        if  height:
            self.change_height(height)
        
    def __repr__ (self):
        return self.name + ": " + ", ".join(["".join([str(num) for num in line]) for line in self.data])
    
    def change_height(self, height):
        self.data = [(e[0] + height ,e[1]) for e in self.data]
        self.height += height
    

class Chamber():

    def __init__(self):
        self.rows = []
        self.heighest_rock = -1

    def find_new_highest(self):
        height = len(self.rows)
        for h in range(self.heighest_rock+1, height):
            if all( map( lambda x: x==VOID, self.rows[h]) ):
                self.heighest_rock = h-1
                break
        return self.heighest_rock
                
    def __repr__(self):
        return "\n".join(["".join(line) for line in self.rows])
    
    def make_room_for(self, rock):
        if dbg: print(f"making room for: {rock}")
        needed = rock[2] + 3
        room_available = len(self.rows) -1- self.heighest_rock

        # rock.height + 3 needs to fit in room_available
        self.rows.extend([VOID for _ in range(7)] for __ in range( needed - room_available))

    def move(self, rock, direction):

        if direction == LEFT:
            new = Rock( ( [ (e[0], e[1]+1) for e in rock.data], rock.name, rock.size ) , 0)
        else:
            assert direction == RIGHT
            new = Rock( ( [ (e[0], e[1]-1) for e in rock.data], rock.name, rock.size ) , 0)
        
        is_valid_move = self.is_valid(new)
        if is_valid_move:
            return is_valid_move , new
        return is_valid_move , rock
    
    def fall(self, rock):
        new = Rock( ( [ (e[0]-1,e[1]) for e in rock.data], rock.name, rock.size ) , 0)
        is_valid_move = self.is_valid(new)
        return is_valid_move , new    

    def is_valid(self, rock):
        
        for row, col in rock.data:
            # cannot move below ground
            if row < 0:
                return False
            
            # cannot move left or right outside boundaries
            if 0 > col or col > 6 :
                return False

            # cannot hit other rocks might have fallen
            if self.rows[row][col] == SOLID:
                return False
        return True
    
    def solidify(self, rock):
        for row, col in rock.data:
            self.rows[row][col] = SOLID
        self.find_new_highest()

def solution_1 (DATA, ROCKS_TO_FALL):
    rocks = cycle( [LINE,PLUS,VERTEX,STICK,SQUARE] )
    gusts = cycle(DATA[0])    
    chamber = Chamber()
    counter = 0
    
    while True:

        rock = next(rocks)
        counter +=1
        if counter % 100000000 == 0:
            print(counter)


        chamber.make_room_for(rock)
        rock = Rock(rock, chamber.heighest_rock +4)
        _, rock = chamber.move(rock, RIGHT)
        _, rock = chamber.move(rock, RIGHT)


        CAN_MOVE = True

        while CAN_MOVE:
            gust = next(gusts)

            CAN_MOVE, moved_rock = chamber.move(rock, gust)

            if CAN_MOVE:
                rock = moved_rock

            CAN_MOVE, fallen_rock = chamber.fall(rock)


            if CAN_MOVE:
                rock = fallen_rock
            else:
                chamber.solidify(rock)
        
        if counter == ROCKS_TO_FALL:
            break
    
    return chamber.find_new_highest() + 1


assert solution_1 (puzzle.test, 2022) == 3068
assert solution_1 (puzzle.data, 2022) == 3179
      
print( f"Solution 1:  {solution_1 (puzzle.data, 2022)} units tall the tower will be after 2022 rocks have stopped falling")

    
```
