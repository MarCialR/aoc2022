```python
def print_image(image):
    from itertools import islice
 
    def chunk(arr_range, arr_size):
        arr_range = iter(arr_range)
        return iter(lambda: tuple(islice(arr_range, arr_size)), ())

    for part in list(chunk(image, 40)):
        print("".join(part))

        
        
class Sprite():
    def __init__(self):
        self.pos = 1
    def add (self, amount):
        self.pos += amount
    def print(self, pos):
        if self.pos-1 <= pos%40  and pos%40 <= self.pos +1:
            return "#"
        return "."
```

```python
dbg = False
CRT_SIZE = 40*6

sprite = Sprite()
image = []

DO_IT_HERE = puzzle.data
it = iter(DO_IT_HERE)

when_to_add = 1
to_add = 1

cycle = 1
instruction = next(it)

try:
    while cycle < CRT_SIZE:
        
        image.append(sprite.print((cycle-1)%40))

        if instruction == "noop":
            instruction = next(it)

        elif instruction.startswith("addx"):
            to_add = int(instruction.split()[1])
            when_to_add = cycle + 1
            instruction = "wait"

        else: # We are actually adding
            if cycle == when_to_add:
                sprite.add(to_add)
                to_add = 0
            instruction = next(it)
        
        cycle += 1 
except StopIteration:
    pass


print( f"Solution 2:\n")
print_image(image)
```
