```python
import re
from collections import deque
testdata="""    [D]    
[N] [C]    
[Z] [M] [P]
 1   2   3 

move 1 from 2 to 1
move 3 from 1 to 3
move 2 from 2 to 1
move 1 from 1 to 2"""

puzzledata = "\n".join(puzzle.data)

RUNTHISDATA = testdata
RUNTHISDATA = puzzledata

def get_numcolumns_linesofdata(data):
    """ returns two pieces of info
    number of stacks
    i -> first line number without stack data"""
    
    for i, line in enumerate(data.splitlines()):
        if line.strip().startswith("1"):
            return int(line.split()[-1]), i

def get_moves(data):
    regex = r"move (\d+) from (\d+) to (\d+)"
    matches = re.finditer(regex, data, re.MULTILINE)
    for matchNum, match in enumerate(matches, start=1):
        yield int(match.group(1)),int(match.group(2)),int( match.group(3))
        

def build_initial_stacks(data, num_columns, lines_of_data ):
    letters = []
    for line in data.splitlines()[:lines_of_data:]:
        letters.append(line[1::4])
    
    stacks = [ deque() for x in range(num_columns)]
    for level in letters[::-1]:
        for i, letter in enumerate(level):
            if letter!= " ":
                stacks[i].append(letter)
    return stacks

numcolumns, linesofdata = get_numcolumns_linesofdata(RUNTHISDATA)  
```

```python
stacks = build_initial_stacks(RUNTHISDATA, numcolumns, linesofdata)

def CrateMover9000 ():
    #print(stacks)
    for repeat, _from, _to in get_moves(RUNTHISDATA):
        for i in range(repeat):
            stacks[_to-1].append(stacks[_from-1].pop())
            #print(stacks)

CrateMover9000()

print ("".join([ s.pop() for s in stacks[0:numcolumns]])) 
```
