```python
from tools import get_puzzle, show_problem_1
TODAY = 18
puzzle = get_puzzle(TODAY)
show_problem_1(puzzle)
```

```python
def solution_1(data):
    
    cubes = set([tuple(map(int,x.split(",") ))for x in data])
    sides = [ (1,0,0),(0,1,0),(0,0,1),(-1,0,0),(0,-1,0),(0,0,-1) ]

    counter = 6*len(cubes)

    for cube in cubes:
        for side in sides:
            if  tuple(sum(t) for t in zip(cube, side)) in cubes:
                counter -=1
    return counter

assert solution_1 (puzzle.test) == 64
      
print( f"Solution 1:  {solution_1 (puzzle.data)} is the surface area of my scanned lava droplet")

    
```

```python
dbg=False
def solution_2(data):
    cubes = set([tuple(map(int,x.split(",") ))for x in data])
    if dbg: print(cubes)
    sides = [ (1,0,0),(0,1,0),(0,0,1),(-1,0,0),(0,-1,0),(0,0,-1) ]
    

    def is_droplet_side_used(cube, side):
        return tuple(sum(t) for t in zip(cube, side)) in cubes    

    def is_empty_side_a_droplet(potential_droplet, droplets):

        if dbg: print(f"\tchecking {potential_droplet}")
        
        if potential_droplet in droplets:
            print("\tAlready determined as droplet !!!!!!!")
            return False
        for side in sides:
            if not is_droplet_side_used(potential_droplet, side):
                if dbg: print(f"\tused in {tuple(sum(t) for t in zip(potential_droplet, side))}")
                return False
        droplets.add(potential_droplet)
        if dbg: print(f"\tempty side is a droplet!\t {droplets}\t {[tuple(sum(t) for t in zip(potential_droplet, side)) for side in sides]}")
        return True

    counter = 6*len(cubes)
    droplets = set()

    for cube in cubes:
        if dbg: print (f"\n================\nchecking cube {cube}")
        for side in sides:
            side = tuple(sum(t) for t in zip(cube, side))
            if side in cubes:
                if dbg: print(f"\n\t{side} is a occupied by cube")
                if dbg: print ("\t-1")
                counter -=1
            else:
                if dbg: print(f"\n\twill now check if {side} is a droplet")
                if is_empty_side_a_droplet(side, droplets):
                    if dbg: print ("\tMINUS 6 !!!")
                    counter -= 6
                
    return counter

assert solution_2 (puzzle.test) == 58

#Solution 2:  3252 is the surface area of my scanned lava droplet   is TOO high
print( f"Solution 2:  {solution_2 (puzzle.data)} is the surface area of my scanned lava droplet")

    
```
