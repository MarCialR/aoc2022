```python
def datapairs(data):
    s_it = iter(solutions.splitlines())
    it = iter(data) 
    while it:
        try:
            l,r = next(it), next(it)
            yield eval(l),eval(r), eval(next(s_it))
            next(it)
        except StopIteration:    #catching the exception
            return

dbg = False

def pre_compare(l,r,l_reminder,r_reminder):
    if dbg:
        print(f"\n\nPREcomp \n=> {l} <=\n=> {r} <=\n///// {l_reminder}\n///// {r_reminder}")
    if type(l) == int and type(r) == int and l==0 and r == 0:
        if dbg:
            print(1)
        return False
    
    failed_l, failed_r = False, False
    try:
        _ = l[0]
    except IndexError:
        failed_l = True
    try:
        _ = r[0]
    except IndexError:
        failed_r = True
    
    if failed_l and failed_r:
        if dbg:
            print(2)
        return pre_compare(l_reminder, l_reminder,0,0)
    
    if failed_l:
        if dbg:
            print(3)
        return True
    if failed_r:
        if dbg:
            print(4)
        return False
    if dbg:
        print(0)
    return compare(l[0],r[0], l[1:] if len(l)>1 else l_reminder, r[1:] if len(r)>1 else r_reminder)

def compare(l,r, l_reminder, r_reminder):
    if dbg:
        print(f"\n\ncomparing \n=> {l} <=\n=> {r} <=\n///// {l_reminder}\n///// {r_reminder}")
    if type(l) == int and type(r) == int:
        if l < r:
            if dbg:
                print(5)
            return True
        elif l > r:
            if dbg:
                print(6)
            return False
        else:
            if dbg:
                print(7)
            return compare (l_reminder,r_reminder,0,0)

    elif type(l) == list and type(r) == list:
        if dbg:
            print(8)
        return compare (l,r,l_reminder, r_reminder)
    else:
        if type(l) == list:
            if dbg:
                print(9)
            return compare(l,[r], l_reminder, r_reminder)
        else:
            if dbg:
                print(10)
            return compare([l],r, l_reminder, r_reminder)

def stephen():
    count = 0
    for i, pair in enumerate(datapairs(puzzle.data)):
        result = pre_compare(pair[0],pair[1],0,0)
        if not result == pair[2]:
            print(f"failed {i+1}  I said {result} but it was {pair[2]}")
            print(f"{pair[0]}")
            print(f"{pair[1]}")
            print(f"==================\n")
        if result:        
            count += i+1
    print (count)

def solve():
    count = 0
    for i, pair in enumerate(datapairs(puzzle.test)):
        result = pre_compare(pair[0],pair[1],0,0)
        print(f"{pair[0]}")
        print(f"{pair[1]}")
        print(f"{result} ==================\n")
        if result:        
            count += i+1
    print (count)

solve()
```

```python
import functools
import itertools
def parse_input(path):
    lines = """[[8, [0], [[1], [1, 9, 4, 5], 1, 5, [2, 3, 8, 8, 10]]], [8], [[[10]], [[3, 5, 3], 5, [6], [6, 10, 7]], [6, [4], [3, 6], [6, 5, 6, 1, 3], [1, 9, 10]]], [[4, [8, 1, 10, 3], [9, 9]]], [6]]
[[[8], 8, [], [4]], [8, [[10, 8, 8], 2, 1, 7, 0], [[], 10, [1, 6, 3], 7], 6], [3], [[8, 10, 9]], [[[], 9, 10, [0, 7, 2, 10, 10]], 10, []]]""".split("\n")
    pairs = [
        list(y) for x, y in itertools.groupby(lines, lambda z: z == "")
        if not x
    ]
    data = [(eval(p1), eval(p2)) for [p1, p2] in pairs]
    return data
def compare_ints(a, b):
    if a < b:
        return -1
    if a == b:
        return 0
    return 1
def compare_two(a, b):
    if type(a) == int and type(b) == int:
        return compare_ints(a, b)
    if type(a) == int:
        return compare_two([a], b)
    if type(b) == int:
        return compare_two(a, [b])
    for (ia, ib) in zip(a, b):
        t = compare_two(ia, ib)
        if t == 0:
            continue
        return t
    return compare_ints(len(a), len(b))
def part1(path):
    data = parse_input(path)
    result = [compare_two(p[0], p[1]) for p in data]
    print(sum([i + 1 for (i, r) in enumerate(result) if r < 0]))
def part2(path):
    data = list(parse_input(path))
    div1 = [[2]]
    div2 = [[6]]
    all_packets = [div1, div2]
    for p in data:
        all_packets.extend(p)
    all_packets.sort(key=functools.cmp_to_key(compare_two))
    idx_div1 = all_packets.index(div1) + 1
    idx_div2 = all_packets.index(div2) + 1
    print(idx_div1 * idx_div2)
```
