```python
from itertools import zip_longest
from collections import deque 

test="""Monkey 0:
  Starting items: 79, 98
  Operation: new = old * 19
  Test: divisible by 23
    If true: throw to monkey 2
    If false: throw to monkey 3

Monkey 1:
  Starting items: 54, 65, 75, 74
  Operation: new = old + 6
  Test: divisible by 19
    If true: throw to monkey 2
    If false: throw to monkey 0

Monkey 2:
  Starting items: 79, 60, 97
  Operation: new = old * old
  Test: divisible by 13
    If true: throw to monkey 1
    If false: throw to monkey 3

Monkey 3:
  Starting items: 74
  Operation: new = old + 3
  Test: divisible by 17
    If true: throw to monkey 0
    If false: throw to monkey 1""".splitlines()

dbg = False


class Monkey():
    
    def __init__(self, monkeys, crunch):
        self.inspection_count = 0
        self.monkeys = monkeys
        
        self.number = int(crunch[0][7:8])
        
        self.items = deque ()
        for x in crunch[1][18:].split(','):
            self.items. append(int(x.strip()))
            
        self.operation = crunch[2][13:]

        self.test = int(crunch[3][21:])
        self.if_true = int(crunch[4][29:])
        self.if_false = int(crunch[5][30:])
        
    def __repr__(self):
        return f"Monkey {self.number} with items: {', '.join([str(x) for x in self.items])}"
    
    def receive_item(self, item):
        self.items.append(item)
    
    def round(self, worry_manage_function):
        if dbg:
            print(str(self) + "\n\n")
        
        while self.items:
            old = self.items.popleft()
            self.inspection_count +=1
            
            if dbg:
                print (f"Monkey inspects an item with a worry level of {old}")
       
            new = 999 # or whatever
            ldict = {"old": old, "new": new}
            exec(self.operation, globals(), ldict)
            new = ldict["new"]
            
            if dbg:
                print (f"Worry level is multiplied by {self.operation} to {new}")
            new = worry_manage_function(new)
            if dbg:
                print (f"Monkey gets bored with item. Worry level is divided by 3 to {new}")
            divisible = (new % self.test) == 0
            if divisible:
                if dbg:
                    print(f"Current worry level is divisible by {self.test}.")
                monkeys[self.if_true].receive_item(new)
                if dbg:
                    print(f"Item with worry level 26 is thrown to monkey {self.if_true}.ºn\n")
            else:
                if dbg:
                    print(f"Current worry level is not divisible by {self.test}.")
                monkeys[self.if_false].receive_item(new)
                if dbg:
                    print(f"Item with worry level 26 is thrown to monkey {self.if_false}.\n\n")
            

```

```python
USE_THIS_DATA = puzzle.data
        
monkeys = [Monkey(monkeys, x) for x in list(zip_longest(*[iter(USE_THIS_DATA)]*7, fillvalue=' ')) ]

worry_function = lambda x: x //3

for i in range(20):
    for m in monkeys:
        m.round(worry_function)

inspections = sorted([m.inspection_count for m in monkeys])
print( f"Solution 1: the two most active monkeys inspected items {inspections[-2]} and {inspections[-1]} times for a total of {inspections[-2] * inspections[-1]}")    
```

```python
import math

USE_THIS_DATA = puzzle.data

monkeys = [Monkey(monkeys, x) for x in list(zip_longest(*[iter(USE_THIS_DATA)]*7, fillvalue=' ')) ]

worry_function = lambda x: x%math.prod([x.test for x in monkeys])

for i in range(10000):
    for m in monkeys:
        m.round(worry_function)

inspections = sorted([m.inspection_count for m in monkeys])
print( f"Solution 2: the two most active monkeys inspected items {inspections[-2]} and {inspections[-1]} times for a total of {inspections[-2] * inspections[-1]}")    
```
