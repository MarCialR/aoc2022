```python
import re

def print_cave (cave):
    with open("cave.out","w") as _file:
        _file.write("\n".join([" ".join([str(p) for p in row])  for row in cave ]) )

def create_map(puzzle):

    _max,_min = max([i for line in S_B_pairs(puzzle) for i in line]), min([i for line in S_B_pairs(puzzle) for i in line])
    print(f"max: {_max}     min: {_min}")
    SIZE = abs (_max * _min)
    cave = [['.' for _ in range(SIZE)] for __ in range(SIZE)]
    
    for line in S_B_pairs(puzzle):
        cave [int(line[1])][int(line[0])] = 'S'
        cave [int(line[3])][int(line[2])] = 'B'
    
    return cave

class Pair():
    def __init__(self, sx,sy,bx,by):
        self.sensor= sx, sy
        self.beacon= bx, by
        self.sx = sx
        self.sy = sy
        self.bx = bx
        self.by = by
        self.closest = abs(sx-bx) + abs(sy-by)
        self.area = self._area()
    def _area (self):
        return ((self.sx - self.closest, self.sx + self.closest), (self.sy - self.closest, self.sy + self.closest))
    
    def safe_area(self, row):
        midpoint = self.sx
        width = self.closest - abs(self.sy-row)
        # we compare against min and max row
        if self.area[1][0] <= row and self.area[1][1] >= row:
            return midpoint - width, midpoint + width
        return []
        
    def __repr__(self):
        return f"S: {self.sensor}\tB: {self.beacon}\tclosest: {self.closest}"

def S_B_pairs(puzzle):
    regex = r"Sensor at x=((\-)?\d+), y=((\-)?\d+): closest beacon is at x=((\-)?\d+), y=((\-)?\d+)"
    values = []
    for line in puzzle:
        match = re.match(regex, line)
        groups = match.groups()
        values.append ( (int(groups[0]),int(groups[2]), int(groups[4]),int(groups[6])) ) 
    return values

#DATA, ROW = puzzle.test, 10
DATA, ROW = puzzle.data, 2000000

pairs_sb = S_B_pairs(DATA)
pairs = [Pair(*pack) for pack in pairs_sb]

cannots = set()
for p in pairs:
    area = p.safe_area(ROW)
    if not area:
        continue
    for i in range(area[0], area[1]+1):
        cannots.add(i)
solution = len(cannots)

#need to remove the Beacons present on row
beacons_counted = set()
for p in pairs:
    if p.by == ROW and (p.bx,p.by) not in beacons_counted:
        beacons_counted.add((p.bx,p.by))
        solution -=1

print(f"Solution: On row {ROW} {solution} positions cannot contain a beacon")
```
