```python
import re

#DATA, ROW, _MAX = puzzle.test, 10, 20
DATA, ROW, _MAX = puzzle.data, 2000000, 4000000


class Pair():
    def __init__(self, sx,sy,bx,by):
        self.sensor= sx, sy
        self.beacon= bx, by
        self.sx = sx
        self.sy = sy
        self.bx = bx
        self.by = by
        self.closest = abs(sx-bx) + abs(sy-by)
        self.area = self._area()
    def _area (self):
        return ((self.sx - self.closest, self.sx + self.closest), (self.sy - self.closest, self.sy + self.closest))
    
    def cannot_row_segment(self, row):
        midpoint = self.sx
        width = self.closest - abs(self.sy-row)
        # we compare against min and max row
        if self.area[1][0] <= row and self.area[1][1] >= row:
            return midpoint - width, midpoint + width
        return []
        
    def __repr__(self):
        return f"S: {self.sensor}\tB: {self.beacon}\tclosest: {self.closest}"

def S_B_pairs(puzzle):
    regex = r"Sensor at x=((\-)?\d+), y=((\-)?\d+): closest beacon is at x=((\-)?\d+), y=((\-)?\d+)"
    values = []
    for line in puzzle:
        match = re.match(regex, line)
        groups = match.groups()
        values.append ( (int(groups[0]),int(groups[2]), int(groups[4]),int(groups[6])) ) 
    return values

pairs_sb = S_B_pairs(DATA)
pairs = [Pair(*pack) for pack in pairs_sb]
def solution_1():
    cannots = set()
    for p in pairs:
        segment = p.cannot_row_segment(ROW)
        if not segment:
            continue
        for i in range(segment[0], segment[1]+1):
            cannots.add(i)
    solution = len(cannots)

    #need to remove the Beacons present on row
    beacons_counted = set()
    for p in pairs:
        if p.by == ROW and (p.bx,p.by) not in beacons_counted:
            beacons_counted.add((p.bx,p.by))
            solution -=1
    return solution

print(f"Solution 1: On row {ROW} there are #  {solution_1()}  # positions that CANNOT contain a beacon")
```

```python
def verify_overlaps(segments):
    i,j = segments[0][0], segments[0][1]
    for seg in segments[1:]:
        if seg[0] > j+1:
            return j+1
        j = max(seg[1], j)
    return None


def solution_2():
    for row in range(_MAX):
        segments = []
        for p in pairs:
            segment = p.cannot_row_segment(row)
            if segment:
                segments. append(segment)
        segments.sort()
        col = verify_overlaps(segments)
        if col:
            print(f"Solution 2: {col*4000000 + row}  (row {row} column  {col})")
            
solution_2()
```
